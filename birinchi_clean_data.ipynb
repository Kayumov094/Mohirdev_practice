Item ustunini toldirish uchun funksiyani yaratib oldim


```python
# Narxlar ro'yxati
prices = [1.0, 2.5, 3.0, 4.0, 5.0, 6.0, 7.0, 8.0, 10.0, 12.0, 14.0, 15.0, 18.0, 20.0]

# Har bir narx uchun Item NaN qiymatlarni to'ldirish
for price in prices:
    # Narxga mos DataFrame yaratish
    category_mode = df[df['Price'] == price].groupby('Category')['Item'].agg(lambda x: x.mode())

    # NaN qiymatlarni moda bilan to'ldirish
    def fill_item_with_mode(row):
        if pd.isna(row['Item']) and row['Price'] == price:
            return category_mode.get(row['Category'], row['Item'])
        return row['Item']

    # DataFrame'da NaN qiymatlarni to'ldirish
    df['Item'] = df.apply(fill_item_with_mode, axis=1)
```
